```python
import csv
import pandas as pd

df = pd.read_csv('train.csv')
df['Age_range'] = df['Age'].apply(lambda a: 1 if a < 25 else 2 if a < 45 else 3)
df.head()

df_x = df[['Pclass','Sex','Age_range','SibSp','Parch','Survived']]
df_x['Sex']=df_x['Sex'].map({'female':1,'male':0})
df_x = df_x.dropna()
X=df_x[['Pclass','Sex','Age_range','SibSp','Parch']]

Y = df_x['Survived']
print X
print Y

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X,Y)

test = pd.read_csv('test.csv')
test['Age_range'] = test['Age'].apply(lambda a: 1 if a < 25 else 2 if a < 45 else 3)
test['SibSp'] = test['SibSp'].apply(lambda a: 8 if a>8 else 8 if a==7 else 5 if a == 6 else a)
test['Parch'] = test['SibSp'].apply(lambda a: 6 if a>6 else a)

test_x = test[['Pclass','Sex','Age_range','SibSp','Parch','PassengerId']]
test_x['Sex']=test_x['Sex'].map({'female':1,'male':0})
test_x = test_x.dropna()
test_x_final = test_x[['Pclass','Sex','Age_range','SibSp','Parch']]

prediction = model.predict(test_x_final)
len(test_x)

test_df =  pd.DataFrame(data = prediction,columns= ['prediction'])

final = pd.DataFrame(data = test_x['PassengerId'], columns = ['PassengerId'])

final['Survived'] = test_df['prediction']
final.head()

final.to_csv('prediction_scikit.csv')
```
