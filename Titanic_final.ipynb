```python
import csv
import pandas as pd

df = pd.read_csv('train.csv')
df['Age_range'] = df['Age'].apply(lambda a: 'youth' if a < 25 else 'medium' if a < 45 else 'old')
#df.head()

survival_freq = pd.value_counts(df.Survived).to_frame().reset_index()
survival_freq.columns=['Survival','freq']
#print survival_freq
prob_survival = survival_freq.apply(lambda x:x/x.sum(),axis=0)
#print prob_survival

class_freq = pd.crosstab(index = df["Survived"], columns = df["Pclass"])
prob_class=class_freq.apply(lambda x: x/x.sum(), axis = 1)
prob_class_0 = prob_class.loc[0]
prob_class_1 = prob_class.loc[1]
#print prob_class_0, prob_class_1
#print prob_class

sex_freq = pd.crosstab(index = df["Survived"], columns = df["Sex"])
prob_sex=sex_freq.apply(lambda x: x/x.sum(), axis = 1)
prob_sex_0 = prob_sex.loc[0]
prob_sex_1 = prob_sex.loc[1]
#print prob_sex_0, prob_sex_1
#print prob_sex

age_freq = pd.crosstab(index = df["Survived"], columns = df["Age_range"])
prob_age=age_freq.apply(lambda x: x/x.sum(), axis = 1)
prob_age_0 = prob_age.loc[0]
prob_age_1 = prob_age.loc[1]
#print prob_age_0, prob_age_1
#print prob_age

sib_freq = pd.crosstab(index = df["Survived"], columns = df["SibSp"])
prob_sib=sib_freq.apply(lambda x: x/x.sum(), axis = 1)
prob_sib_0 = prob_sib.loc[0]
prob_sib_1 = prob_sib.loc[1]
#print prob_sib_0, prob_sib_1
#print prob_sib

par_freq = pd.crosstab(index = df["Survived"], columns = df["Parch"])
prob_par=par_freq.apply(lambda x: x/x.sum(), axis = 1)
prob_par_0 = prob_par.loc[0]
prob_par_1 = prob_par.loc[1]
#print prob_par_0, prob_par_1
#print prob_par

prob_loc_0 = {}
count1 = 0
for a in prob_class_0:
    count1 = count1 + 1 
    count2 = 0
    for b in prob_sex_0:
        count2 = count2 + 1
        count3 = 0
        for c in prob_age_0:
            count3 = count3+1
            count4 = 0
            for d in prob_sib_0:
                count4 = count4+1
                count5 = 0   
                for e in prob_par_0:        
                    count5 = count5+1
                    prob_loc_0[str(count1)+str(count2)+str(count3)+str(count4)+str(count5)] = a*b*c*d*e*prob_survival.loc[0,'freq']

#print prob_loc_0

prob_loc_1 = {}
count1 = 0
for a in prob_class_1:
    count1 = count1 + 1 
    count2 = 0
    for b in prob_sex_1:
        count2 = count2 + 1
        count3 = 0
        for c in prob_age_1:
            count3 = count3+1
            count4 = 0
            for d in prob_sib_1:
                count4 = count4+1
                count5 = 0   
                for e in prob_par_1:        
                    count5 = count5+1
                    prob_loc_1[str(count1)+str(count2)+str(count3)+str(count4)+str(count5)] = a*b*c*d*e*prob_survival.loc[1,'freq']

#print prob_loc_1

prob_loc = {}
for a in prob_loc_0:
    if prob_loc_0[a] > prob_loc_1[a]:
        prob_loc[a]=0
    else:
        prob_loc[a]=1
#print prob_loc

columns = [list(class_freq.columns.values), list(sex_freq.columns.values), list(age_freq.columns.values), list(sib_freq.columns.values),list(par_freq.columns.values) ]
#print columns

test = pd.read_csv('test.csv')
test['Age_range'] = test['Age'].apply(lambda a: 'youth' if a < 25 else 'medium' if a < 45 else 'old')
test['SibSp'] = test['SibSp'].apply(lambda a: 8 if a>8 else 8 if a==7 else 5 if a == 6 else a)
test['Parch'] = test['SibSp'].apply(lambda a: 6 if a>6 else a)
#test.head()

test_cols = list(test.columns.values)
test_cols.insert(0,'S.No.')
test_cols.append('Attributes')
test_cols.append('Prediction')
#print test_cols
df_predict = []
for a in test.itertuples():
    temp = list(a)
    x = str(columns[0].index(a[2])+1)+str(columns[1].index(a[4])+1)+str(columns[2].index(a[12])+1)+str(columns[3].index(a[6])+1)+str(columns[4].index(a[7])+1)
    prediction = prob_loc[x]
    temp.append(x)
    temp.append(prediction)
    df_predict.append(temp)
final_predict = pd.DataFrame(df_predict,columns = test_cols)
del final_predict['S.No.']
#final_predict

final_predict.to_csv('prediction.csv')
```
